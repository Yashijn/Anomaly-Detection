```python
from bs4 import BeautifulSoup
import requests
import time
from datetime import timedelta
from requests.adapters import HTTPAdapter
from urllib.request import Request,urlopen
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
urls=[]
titles=[]
descriptions=[]
time=[]
```

```python
for l in link:
  req=Request(l,headers={'Referer': 'https://www.moneycontrol.com','User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'})
  webpage=urlopen(req).read()

  with requests.Session() as c:
    soup=BeautifulSoup(webpage,'html5lib')
    #print(soup)
    for item in soup.find_all('div',attrs={'class':'FL PR20'}):
      raw_link=(item.find('a',href=True)['href'])
      link='https://www.moneycontrol.com'+raw_link
      urls.append(link)
      print(link)
      title=(item.find('a',href=True)['title'])
      print(title)
```

```python
for i in urls:
  req=Request(i,headers={'Referer': 'https://www.moneycontrol.com','User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'})
  webpage=urlopen(req).read()

  with requests.Session() as c:
    soup=BeautifulSoup(webpage,'html5lib')
    #print(soup)
    print(i)
    title=soup.find('h1',attrs={'class':'article_title artTitle'})
    title=title.text
    title=title.replace(",","")
    titles.append(title)
    print(title)
    desc=soup.find('h2',attrs={'class':'article_desc'})
    desc=desc.text
    desc=desc.replace(",","")
    descriptions.append(desc)
    print(desc)
    timings=soup.find('div',attrs={'class':'article_schedule'})
    timings=timings.text
    timings=timings.split('/')[0]
    timings=timings.replace(",","")
    time.append(timings)
    print(timings)
```

```python
import datetime
corrected_dates=[]
for i in time:
  modified_date=datetime.datetime.strptime(i,"   %B %d %Y  ")
  corrected_dates.append(modified_date)
```

```python
import pandas as pd
df=pd.DataFrame(list(zip(corrected_dates, titles, descriptions,urls)), columns =['Date', 'Title', 'Description','link'])
df=df.sort_values('Date')
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#Adding some new words to Vader Dictionary to judge stock market news better.

new_words =  {'falls': -9, 'drops': -9, 'rise': 9, 'increases': 9, 'gain': 9, 'hiked': -9, 'dips': -9, 'declines': -9, 'decline': -9, 'hikes': -9, 'jumps': 9,
              'lose': -9, 'profit': 9, 'loss': -9, 'shreds': -9, 'sell': -9, 'buy': 9, 'recession': -9, 'rupee weakens': -9, 'record low': -9, 'record high': 9,
              'sensex up': 9, 'nifty down': -9, 'sensex down': -9, 'nifty up': 9,'high':9,'gains':9,'surges':9,'jump':9,'wrongdoing':-9,'down':-9,'up':9,'refuses':-9,'wins':9,'Consolidated':9} 

analyser = SentimentIntensityAnalyzer()
analyser.lexicon.update(new_words)
```

```python
f=lambda title:analyser.polarity_scores(title)['compound']
df['score']=df['Title'].apply(f)
```

```python
import yfinance as yf
df_stock= yf.download("ADANIENT.NS", start="2020-01-01", end="2021-05-20",group_by="ticker") 

df_stock['effect 2 day before']=df_stock['Close'].shift(1)-df_stock['Close'].shift(2)
df_stock['effect 1 day before']=df_stock['Close']-df_stock['Close'].shift(1)
df_stock['effect on the day']=df_stock['Close']-df_stock['Open']
df_stock['effect 1 day after']=df_stock['Close'].shift(-1)-df_stock['Close']
df_stock['effect 2 day after']=df_stock['Close'].shift(-2)-df_stock['Close'].shift(-1)
df_stock['effect 3 day after']=df_stock['Close'].shift(-3)-df_stock['Close'].shift(-2)
df_stock['effect 4 day after']=df_stock['Close'].shift(-4)-df_stock['Close'].shift(-3)
df_stock['effect 5 day after']=df_stock['Close'].shift(-5)-df_stock['Close'].shift(-4)
df_stock.drop('Adj Close',axis=1,inplace=True)
df_stock.drop('Open',axis=1,inplace=True)
df_stock.drop('High',axis=1,inplace=True)
df_stock.drop('Low',axis=1,inplace=True)
df_stock.drop('Volume',axis=1,inplace=True)
#df_stock.drop('Next_day_Close',axis=1,inplace=True)

```

```python
#df_stock['Date'] = pd.to_datetime(df_stock.Date, format='%Y-%m-%d')
df.drop('link',axis=1,inplace=True)
df_combo=pd.merge(df,df_stock,how='inner',on='Date')
df_combo
```

Analysing the dataframe the following inferences can be made----

*   The effect of the news on 2020-01-20 was seen 1 day before and after the release of the news and also on the date of release of the news.
*   Despite of the positive compound score given by vader on the news dated on 2020-03-05	and 2020-03-11 ,the stock prices decreased.In fact, in the case of news on 2020-03-11 the 'momentum effect' can be seen.
*   The effect of news on 2021-03-23 was witnessed before the time of its release.
*   The effect of the news on 2020-08-25 was seen 1 day after its release.
*   THe score given by Vader lexicon on 2020-08-06, 2020-08-07, 2020-11-06,2021-04-16,2021-03-23 couldn't justify the fluctuation in prices

Hence it can be inferred that Vader lexicon is not much helpful in predicting the effect of news on stock price movement
